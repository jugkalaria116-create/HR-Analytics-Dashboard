```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("../data/IBM HR Employee Attrition.csv")
df.head()
print(df.shape)
print(df.columns)
print(df.info())
```

Missing Values Check

```python
df.isnull().sum()
```

Total Employees

```python
print("Total Employees:", len(df))
```

Attrition Count

```python
print(df["Attrition"].value_counts())
```

Attrition %

```python
attrition_rate = (df["Attrition"].value_counts()["Yes"] / len(df)) * 100
print("Attrition Rate:", round(attrition_rate,2), "%")
```

Attrition by Department

```python
df.groupby("Department")["Attrition"].value_counts()
```

Average Salary by Department

```python
df.groupby("Department")["MonthlyIncome"].mean().round(2)
```

Gender Count

```python
df["Gender"].value_counts()
```

Overtime vs Attrition

```python
pd.crosstab(df["OverTime"], df["Attrition"])
```

Attrition Count Chart

```python
df["Attrition"].value_counts().plot(kind="bar")
plt.title("Attrition Count")
plt.show()
```

Department Wise Employees

```python
df["Department"].value_counts().plot(kind="bar")
plt.title("Employees by Department")
plt.show()
```

Gender Ratio Pie Chart

```python
df["Gender"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Gender Distribution")
plt.ylabel("")
plt.show()
```
